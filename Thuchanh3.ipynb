```python
from filterpy.kalman import KalmanFilter
import numpy as np

# Function to apply Kalman Filter
def apply_kalman_filter(data):
    kf = KalmanFilter(dim_x=2, dim_z=1)
    kf.x = np.array([0., 0.])       # initial state (location and velocity)
    kf.F = np.array([[1., 1.], [0., 1.]]) # state transition matrix
    kf.H = np.array([[1., 0.]])     # Measurement function
    kf.P *= 1000.                   # covariance matrix
    kf.R = 5                        # state uncertainty
    kf.Q = np.array([[0.1, 0.1], [0.1, 0.1]]) # process uncertainty

    filtered_data = []
    for z in data:
        kf.predict()
        kf.update(z)
        filtered_data.append(kf.x[0])
    return filtered_data

# Áp dụng Kalman Filter cho cột thứ 13 (Feature_13)
filtered_feature_13 = apply_kalman_filter(X[:, 0])

# So sánh dữ liệu gốc và dữ liệu đã lọc
plt.figure(figsize=(10, 6))
plt.plot(X[:, 0], label='Original ' + df.columns[13])
plt.plot(filtered_feature_13, label='Filtered ' + df.columns[13], linestyle='dashed')
plt.title('Kalman Filter on ' + df.columns[13])
plt.xlabel('Index')
plt.ylabel(df.columns[13])
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans

# Tạo một DataFrame mới với cột 13, cột 14 đã được lọc, và cột 15
df_filtered = pd.DataFrame({
    df.columns[13]: X[:, 0],
    df.columns[14]: filtered_feature_14,
    df.columns[15]: X[:, 2]
})

# Sử dụng k-means clustering với 3 cụm
kmeans = KMeans(n_clusters=3, random_state=42)
df_filtered['Cluster'] = kmeans.fit_predict(df_filtered)

# Trực quan hóa kết quả cụm
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_filtered[df.columns[14]], y=df_filtered[df.columns[15]], hue=df_filtered['Cluster'], palette='viridis')
plt.title('K-means Clustering on Filtered Data')
plt.xlabel(df.columns[13])
plt.ylabel(df.columns[14])
plt.show()
```
